```python
data = pd.read_csv('book.csv', encoding = 'Latin1')
data = data.rename(columns = {'User.ID':'UserId', 'Book.Title':'Booktitle', 'Book.Rating':'Bookrating'})
data.head()
```

```python
len(data.Booktitle.unique())
```

```python
len(data.UserId.unique())
```

```python
data1 = data.pivot_table(index = 'UserId', columns = "Booktitle", values = "Bookrating", aggfunc = sum)
data1
```

```python
data1.index = data.UserId.unique()
data1
```

```python
data1.fillna(0, inplace=True)
data1
```

```python
data_user = 1-pairwise_distances(data1.values,metric="cosine")
data_user
```

```python
data_user_df = pd.DataFrame(data_user)
```

```python
data_user_df.index = data.UserId.unique()
data_user_df.columns = data.UserId.unique()
```

```python
data_user_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(data_user,0)
data_user_df.iloc[0:5,0:5]
```
